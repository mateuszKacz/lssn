```python
a=0.7
b=0.8
τ=12.5
```

```python
def fhn(X,t):
    v,w=X
    dX=zeros(2)
    dX[0] = v-v**3/3-w+Iext(t)
    dX[1] = (v + a - b*w)/τ
    return dX

def Iext(t):
    return 0.5
```

```python
T = np.linspace(0, 100, 10000)
x0=array([0.1, 0.2])

x = odeint(fhn, x0, T)
```

```python
subplots(figsize=(15, 5))
subplot(121)
plot(x[:,0],x[:,1])
subplot(122)
plot(T,x[:,0]);
```
